## Univariate Analysis

*Histogram plots of numerical data*

```python
df.hist(figsize=(20,20));
```

*What is the distribution of customer age?*

```python
plt.style.use("default") # using default style

df['Age'].hist() # histogram plot of age

# setting title and axis labels
plt.title('Number of Customers by Age')
plt.ylabel('Number of Customers')
plt.xlabel('Age')

plt.grid(False) # removing grid lines

plt.show()
```

*How many customers complained?*

```python
plt.style.use("ggplot") # using style ggplot

ax = df['Complain'].value_counts().plot(kind='barh') # horizontal bar chart

# setting title and axis labels
plt.title('Number of Customer Complaints')
plt.xlabel('Number of Customers')
plt.ylabel('Complaint 0=No, 1=Yes')

# annotate
ax.bar_label(ax.containers[0], label_type='edge')

# pad the spacing between the number and the edge of the figure
ax.margins(x=0.1)

# inverting y axis to show 0 at top
plt.gca().invert_yaxis()

plt.show()
```

*Which marketing campaign was the most successful?*

```python
# creating a list of campaign column headers
campaigns = ['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'Response']

# campaign results to dataframe
results = pd.DataFrame({"No. of People Accepted": df[campaigns].sum(), 
                       "Percentage No. Accepted": df[campaigns].mean().round(4)*100})

# campaign success rates output
results
```

```python
# horizontal bar chart
df[campaigns].sum().plot(kind='barh') 

# setting title and axis labels
plt.title('Which marketing campaign was the most successful?')
plt.xlabel('Number of Customers that Accepted the Campaign Offer')
plt.ylabel('Campaign')

plt.show()
```

*Which products are performing best?*

```python
# making a list of the different product areas
product = ['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']

# horizontal bar chart of total values and sorted highest to lowest
df[product].sum().sort_values().plot(kind='barh', color='red')

# setting title and axis labels
plt.title('Which products are performing best?')
plt.xlabel('Total amount spent')
plt.ylabel('Product')

plt.show()
```

*Which channels are underperforming?*

```python
# making a list of the different business channels
channel = ['NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases']

# horizontal bar chart of total values and sorted highest to lowest
df[channel].sum().sort_values().plot(kind='barh', color='green')

# setting title and axis labels
plt.title('Which business channels are performing best?')
plt.xlabel('Total number of purchases')
plt.ylabel('Business channel')

plt.show()

display(df[channel].sum().sort_values(ascending=False))
```

<div class="alert alert-block alert-info">

### Insights from univariate analysis
    
* There are 3 customers identified as outliers in the Age column, older than 120 years
* Very low number of customer complaints, at less than 1%
* ‚ÄúResponse Campaign‚Äù which is the last marketing campaign is the most successful, with over 15% of customers taking the offer. ‚ÄúCampaign 2‚Äù was the least performing campaign, with just 1.35% success.
* Wine has the highest consumption when compared with other products, with over 80% more revenue than the next closest product Meat.
* Customers are most interested to buy products directly from stores as opposed to any other channel. When we look at just the number of purchases, deals and catalog sales seem to be underperforming. But ideally, we would need to calculate the return of investment for each of the channels to accurately assess the performance.

</div>   

## Bivariate Analysis

*What factors are significantly related to the number of web purchases?*

```python
# correlations based on web purchases
corr = df.corr()[['NumWebPurchases']].sort_values(by='NumWebPurchases', ascending=False)

# plot size
plt.figure(figsize=(2, 6))

# correlation heatmap and title
heatmap = sns.heatmap(corr, annot=True, cmap='Greens')
heatmap.set_title('The most linear correlated features to NUMBER OF WEB PURCHASES', 
                  fontdict={'fontsize':18}, pad=16);
```

```python
print('Web purchases by education level')

# required outputs to dataframe
Education = pd.DataFrame({"No. of customers": df["Education"].value_counts(), 
                       "No. of web purchases": df.groupby("Education")["NumWebPurchases"].sum(),
                          "Avg. web purchases": df.groupby("Education")["NumWebPurchases"].mean().round(1)})

# output
Education.sort_values('Avg. web purchases', ascending=False)
```

```python
print('Web purchases by marital status')

# required outputs to dataframe
Marital = pd.DataFrame({"No. of customers": df["Marital_Status"].value_counts(), 
                       "No. of web purchases": df.groupby("Marital_Status")["NumWebPurchases"].sum(),
                          "Avg. web purchases": df.groupby("Marital_Status")["NumWebPurchases"].mean().round(1)})

# output
Marital.sort_values('Avg. web purchases', ascending=False)
```

```python
print('Web visits per month by education level')

# required outputs to dataframe
Education2 = pd.DataFrame({"No. of customers": df["Education"].value_counts(), 
                       "No. of monthly web visits": df.groupby("Education")["NumWebVisitsMonth"].sum(),
                          "Avg. web visits/month": df.groupby("Education")["NumWebVisitsMonth"].mean().round(1),
                          "No. of web purchases": df.groupby("Education")["NumWebPurchases"].sum()})

# output
Education2.sort_values('Avg. web visits/month', ascending=False)
```

*Amount spent on various products by Country*

```python
df.groupby("Country")[product].sum()
```

```python
df['Total_Spend'] = df['MntWines'] + df['MntFruits'] + df['MntMeatProducts'] + df['MntFishProducts'] + df['MntSweetProducts'] + df['MntGoldProds']

df.head()
```

```python
plt.style.use("default")

df.groupby("Country")['Total_Spend'].sum().plot(kind='pie', autopct='%1.1f%%', pctdistance=0.7, 
                                                figsize=(10, 10))

plt.title('Percentage of Total Spend by Country')

plt.show()
```

<div class="alert alert-block alert-info">

### Insights from bivariate analysis
    
* Education-level is positively correlated with the number of web purchases. People with higher education tend to do more web purchases.
* Wine products are most strongly correlated with web purchases, with corr value +0.55
* Web visits have a very weak negative correlation with web purchases. People with lower education levels tend to spend more time on the website but they do not result in higher purchases.
* Spain is the highest represented country in the data and contributes to nearly 50% of the total product spend when compared to other countries. 

</div>

# <font color = darkred> Marketing Recommendations </font>

```python
print('Analysis of total spend by country')

# required outputs to dataframe
Spend = pd.DataFrame({"No. of customers": df["Country"].value_counts(), 
                       "Total spend": df.groupby("Country")["Total_Spend"].sum(),
                          "Avg. spend per capita": df.groupby("Country")["Total_Spend"].mean().round(1),
                          "Avg. Income": df.groupby("Country")["Income"].mean().round(1)})

# output
Spend.sort_values('Total spend', ascending=False)
```

```python
# correlations based on total spend
corr1 = df.corr()[['Total_Spend']].sort_values(by='Total_Spend', ascending=False)

# plot size
plt.figure(figsize=(2, 6))

# correlation heatmap and title
heatmap = sns.heatmap(corr1, annot=True, cmap='Greens')
heatmap.set_title('The most linear correlated features to TOTAL SPEND', 
                  fontdict={'fontsize':18}, pad=16);
```

<div class="alert alert-block alert-info">

**Marketing Recommendations:**
    
* Focus more on Wine and Meat because countries are spending more on these products
* Look at ways to increase web purchases as these are low cost revenue streams. Perhaps direct marketing towards PhD and Graduate level educated customers who offer more web purchases
* Focus on countries with higher average Income, where the average spend per capita is more, which would be Mexico, Saudi Arabia.
* Try to develop business in Mexico and USA further, as these markets offer high average revenue and return, but are currently areas where we have the least amount of customers
